## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
```

```python
data = pd.read_csv('../weather-raw.csv')
data
```

```python
data.columns
```

```python
data.dtypes
```

```python
# elijiremos columnas que están totalmente vacías
empty_colum = data.isnull().sum()
empty_colum[empty_colum == 22]
```

```python
delete_colum = list(empty_colum[empty_colum == 22].index)
print(delete_colum)
```

```python
# aqu+í las eliminamos
data = data.drop(delete_colum, axis =1)
data
```

```python
# para que no haya valores vacíos, vamos a llenar con valores 'cero'
data = data.fillna(0)
data
```

```python
# obtenemos las columnas ya modificadas, para ocuparla en el pivoteo de columnas
list_col = data.columns[4:]
list_col
```

```python
# con esta visualización podemos ver la temperatura priorizando por está misma
# a mi parecer es mejor esta visualización
df_by_month= data.pivot(index='month', columns='element',values = list_col)
df_by_month
```

```python
# con esta visualización podemos ver la temperatura, priorizando por día
df_by_temp= data.pivot(index='element', columns='month',values = list_col)
df_by_temp
```
